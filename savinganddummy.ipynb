```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression # linear regression
model=LinearRegression() # stored function in object
from sklearn.model_selection import train_test_split #testing and train
from sklearn.preprocessing import LabelEncoder  # another method dummy labelencoder they arrange name 
le=LabelEncoder()
from sklearn.preprocessing import OneHotEncoder
```

```python

Age=['6','3','5','2','4','5','5','6','8','6','7','7']
Mileage=['69000','35000','57000','22500','46000','59000','52000','72000','91000','67000','83000','79000']
Price=['18000','34000','26100','40000','31500','29400','32000','19300','12000','22000','20000','21000']
Car_Model=['BMW X5','BMW X5','BMW X5','BMW X5','BMW X5','Audi A5','Audi A5','Audi A5','Audi A5','Mercedez','Mercedez','Mercedez']
```

```python
df=pd.DataFrame({'carmodel':Car_Model,'mileage':Mileage,'price':Price,'age':Age})
df
```

```python
dummy_var=pd.get_dummies(df.carmodel)
dummy_var
```

```python
merged=pd.concat([df,dummy_var],axis='columns')
merged
```

```python
final=merged.drop(['carmodel','Mercedez'],axis='columns')
final
```

```python
x=final.drop('price',axis='columns')
y=final.price
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
model=model.fit(x_train,y_train)
```

```python
x_test
```

```python
pred=model.predict(x_test)
pred
```

```python
model.predict([[59000,5,1,0]])
```

```python
model.score(x_train,y_train)
```

```python
# sklearnpreprocessing one hot encoder

dd=df
dd.carmodel=le.fit_transform(dd.carmodel)
dd
```

```python
x1=dd[['carmodel','mileage','age']]
y1=dd.price
```

```python
# another method one hot encoder

from sklearn.compose import ColumnTransformer
oye = ColumnTransformer([('carmodel', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
x2=oye.fit_transform(x1)
x2
```

```python
x2=x2[:,1:]
x2
```

```python
model.fit(x2,y1)
model.score(x2,y1)
```

```python
model.predict([[0, 0,59000, 5]])
```
